```python
import pandas as pd
#コーヒーサイズデータフレーム作成
df = pd.DataFrame({
    'SIZE': ['L', 'M', 'S', 'L'],
    'PRICE': ['250', '200', '150', '250']
})

#表示
df
```

```python
#category encoder インポート
import category_encoders as ce
#マップを辞書で学習
mapping = [{'col': 'SIZE', 'mapping': {'S': 0, 'M': 1, 'L': 2}}]
#エンコーダーをインスタンス化
enc = ce.OrdinalEncoder(mapping=mapping)
#マップを学習
enc.fit(df)
#変換
enc.transform(df)
```
