```python
fonts = set([f.name for f in matplotlib.font_manager.fontManager.ttflist])
print(fonts)
```

```python
# 
plt.figure(figsize=(10,len(fonts)/4))
 
# 
for i, font in enumerate(fonts):
    plt.text(0, i, f"{font}", fontname=font)
    # 
plt.ylim(0, len(fonts))
plt.axis("off")
    
plt.show()
```
